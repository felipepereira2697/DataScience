## Create new variables

```python
import pandas as pd
```

```python
dataset = pd.read_csv('data/rental_residence.csv',sep=';')
```

```python
dataset.head(10)
```

```python
# creating new value Total --> valor+condominio_iptu
dataset['Valor Bruto'] = dataset['Valor'] + dataset['Condominio'] + dataset['IPTU']
```

```python
dataset
```

```python
dataset['Valor m2'] = (dataset['Valor'] / dataset['Area']).round(2)
```

```python
dataset
```

```python
dataset['Valor Bruto m2'] = (dataset['Valor Bruto'] / dataset['Area']).round(2)
```

```python
dataset
```

```python
houseType = ['Casa','Casa de Condomínio', 'Casa de Vila']
```

```python
#apply function executes for each record, row
dataset['Aggregate Type'] = dataset['Tipo'].apply(lambda x: 'Casa' if x in houseType else 'Apartamento')
```
