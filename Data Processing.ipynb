## Data Processing

```python
#Missing Data
import pandas as pd
data = pd.read_csv('data/rental_residence.csv',sep=';')
data.head(10)
```

```python
data.isnull()
```

```python
data.info()
```

```python
#create a boolean series and check true if is null
data['Valor'].isnull()
```

```python
#These are the records that contains Valor equals null
data[data['Valor'].isnull()]
```

```python
A = data.shape[0]
```

```python
data.dropna(subset=['Valor'], inplace=True)
B = data.shape[0]
A - B
```

```python
#To double check
#These are the records that contains Valor equals null the return is a null dataframe
data[data['Valor'].isnull()]
```

```python
data[data['Condominio'].isnull()].shape[0]
```

```python
#apartamento com condominio nulo
select = (data['Condominio'].isnull()) & (data['Tipo'] == 'Apartamento')
data[select]
```

```python
A = data.shape[0]
#using ˜select we change false to true and true to false
data = data[~select]
B = data.shape[0]
A-B
```

```python
#we can do like this
#data.fillna(0,inplace=True)
data = data.fillna({'Condominio':0, 'IPTU':0})
```

```python
data.to_csv('data/rental_residence.csv',sep=';', index=False)
```
