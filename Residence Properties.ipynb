## Residence Properties Report

```python
import pandas as pd
```

```python
data = pd.read_csv('data/aluguel.csv', sep=';')
data.head(10)
list(data['Tipo'].drop_duplicates())
```

```python
residences = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio'
 'Casa de Vila']
```

```python
type(residences)
```

```python
#isin method check on the columns Tipo if we have any occurrences of at least one of the items that is inserted
#on residences list --> isin returns a boolean series
selection = data['Tipo'].isin(residences)
```

```python
#filtering data with selection
residence_data = data[selection]
residence_data
```

```python
# To check if we have non residence types
list(residence_data['Tipo'].drop_duplicates())
```
